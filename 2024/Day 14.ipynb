##### --- Day 14: Restroom Redoubt ---

--- Part One --- + --- Part Two ---

```python
import re
import matplotlib.pyplot as plt
import numpy as np

file = open("day_14.txt", "r")
the_input = [i.strip() for i in file.readlines()]

grid_size = (101, 103)
robots = []
for ind, line in enumerate(the_input):
    robots.append([int(x) for x in re.findall('[0-9]+|-[0-9]+', line)])
    
for i in range(10000):
    for r in robots:
        new_pos = (r[0] + r[2]) % grid_size[0] , (r[1] + r[3]) % grid_size[1]
        r[0] = new_pos[0]
        r[1] = new_pos[1]
    if i + 1 == 100:
        # Calculate and print the safety score.
        half_h = grid_size[1]//2
        half_w = grid_size[0]//2
        q_1 = q_2 = q_3 = q_4 = 0
        for r in robots:
            x = r[0]
            y = r[1]
            if x < half_w and y < half_h:
                q_1 += 1
            if x < half_w and y > half_h:
                q_3 += 1
            if x > half_w and y < half_h:
                q_2 += 1
            if x > half_w and y > half_h:
                q_4 += 1
        print(q_1 * q_2 * q_3 * q_4)
        
    if i == 6875:
        print(f"After {i+1} seconds:")
        x = np.array([r[0] for r in robots])
        y = np.array([r[1] for r in robots])
        plt.scatter(x, y, s=10, c=np.random.uniform(20, 80, len(x)), vmin=0, vmax=100)
        plt.show()
```
